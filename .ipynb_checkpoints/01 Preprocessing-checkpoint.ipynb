```python
#import library 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import CountVectorizer

import string
import re
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
import swifter
import emoji
import contractions

import nltk
from nltk.metrics.distance import edit_distance
from nltk.probability import FreqDist
from nltk.tokenize import word_tokenize 
from nltk.corpus import stopwords
nltk.download('stopwords')
nltk.download('punkt')

from wordcloud import WordCloud
```

```python
#df = pd.read_csv("ricis_official.csv")
#df = pd.read_csv(".01 DATA_TA/ricis_official.csv")
#df = pd.read_csv(".01 DATA_TA/raw_data/ken_and_grat.csv")
#df = pd.read_csv(".01 DATA_TA/raw_data/gadgetIn.csv")
#df = pd.read_csv(".01 DATA_TA/raw_data/go_indonesia.csv")
df = pd.read_csv(".01 DATA_TA/raw_data/tereza_fahlevi.csv")
```

```python
#Hapus atribut yang tidak digunakan
#df.drop(["web-scraper-order","web-scraper-start-url","video_link","video_link-href","judul","likes","komen"], axis = 1, inplace = True)
df.drop(["web-scraper-order","web-scraper-start-url","vidio_link","vidio_link-href","judul","likes","komen"], axis = 1, inplace = True)
#ganti atribut yang tidak digunakan
df.rename({"teks_komen": "comments"}, axis = 1, inplace = True)
```

```python
print(df.shape)
df.head(10)
```

```python
#Case Folding
df['comments'] = df['comments'].str.lower()
```

```python
#Cleansing Text
def cleaning_text(text):
    # menghapus RT tag
    text = re.sub('RT\s', '', str(text))
    # menghapus @_username
    text = re.sub('\B@\w+', '', text)
    # menghapus emoji dengan text
    text = emoji.demojize(text)
    # menghapus URL
    text = re.sub('(http|https):\/\/\S+', '', text)
    # menghapus #_something_
    text = re.sub('#+', '', text)
    # konversi huruf kapital ke huruf kecil semua
    text = text.lower()
    # menghapus kata yang berulang-ulang ('oooooo' menjadi '00')
    text = re.sub(r'(.)\1+', r'\1\1', text)
    # menghapus punctuation repetition dengan single occurance ('!!!!!!' ,enjadi '!')
    text = re.sub(r'[\?\.\!]+(?=[\?.\!])', '', text)
    # menghilangkan angka dan spesial karakter, hanya mengambil alfabet saja
    text = re.sub(r'[^a-zA-Z]', ' ', text)
    # menghapus contractions 
    text = contractions.fix(text)
    return text

df['comments'] = df['comments'].apply(lambda x: cleaning_text(x))
df.head(10)
```

```python
#Tokenization
def tokenization(text):
    text = re.split('\W+', text)
    return text

df['Tokenization'] = df['comments'].apply(lambda x: tokenization(x.lower()))
df.head(10)
```

```python
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
kalimat = "ada siapa di mana hehe gagag tata"

tokens = nltk.tokenize.word_tokenize(kalimat)
kemunculan = nltk.FreqDist(tokens)
print(kemunculan.most_common())
```

```python
#Stop Removal
#menghilangkan kata kurang efektif
stopword = nltk.corpus.stopwords.words('indonesian')

def remove_stopwords(text):
    text = [word for word in text if word not in stopword]
    return text

df['Stop_Removal'] = df['Tokenization'].apply(lambda x: remove_stopwords(x))
df.head(10)
```

```python
# STEMMING
# create stemmer
factory = StemmerFactory()
stemmer = factory.create_stemmer()

# stemmed
def stemmed_wrapper(term):
    return stemmer.stem(term)

term_dict = {}

for document in df['Stop_Removal']:
    for term in document:
        if term not in term_dict:
            term_dict[term] = ' '
            
print(len(term_dict))
print("------------------------")

for term in term_dict:
    term_dict[term] = stemmed_wrapper(term)
    print(term,":" ,term_dict[term])
    
print(term_dict)
print("------------------------")


# apply stemmed term to dataframe
def get_stemmed_term(document):
    return [term_dict[term] for term in document]

df['Comments_Stemmed'] = df['Stop_Removal'].swifter.apply(get_stemmed_term)
print(df['Comments_Stemmed'])
```

```python
#WordCloud komen
stop_words = set(stopwords.words('indonesian'))
df['title_no_stopwords'] = df['comments'].apply(lambda x: [item for item in str(x).split() if item not in stop_words])

all_words = list([a for b in df['title_no_stopwords'].tolist() for a in b])
all_words_str = ' '.join(all_words) 

def plot_cloud(wordcloud):
    plt.figure(figsize=(30, 20))
    plt.imshow(wordcloud) 
    plt.axis("off");

wordcloud = WordCloud(width = 2000, height = 1000, random_state=1, background_color='white', 
                      colormap='viridis', collocations=False).generate(all_words_str)
plot_cloud(wordcloud)
```
